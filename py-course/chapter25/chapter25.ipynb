```python
df_data = pd.read_csv("data.csv")
df_data
```

```python
df_data = df_data.loc[(df_data.UnitPrice > 0) & (df_data.UnitPrice < 50)]
df_data = df_data.loc[(df_data.Quantity > 0) & (df_data.Quantity < 800)]
df_data.InvoiceDate = pd.to_datetime(df_data.InvoiceDate)
df_data = df_data.set_index("InvoiceDate")
df_data

```

```python
df_data["date"] = df_data.index.date
df_data
```

```python
df_data["TotalPrice"] = df_data["UnitPrice"] * df_data["Quantity"]
df_data
```

```python
df_cum = df_data[["TotalPrice", "date"]]
df_cum
```

```python
df_cum = df_cum.groupby("date").sum()
df_cum
```

```python
trace = go.Scatter(x=df_cum.index.astype(str), y=df_cum.TotalPrice,
                    opacity = 0.8, line = dict(color = '#38C788'), name= 'Transcation by day')
layout = dict(
    title= "Informations by Date",
    paper_bgcolor='#ffffff',
    template='plotly_white',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label='1m', step='month', stepmode='backward'),
                dict(count=3, label='3m', step='month', stepmode='backward'),
                dict(count=6, label='6m', step='month', stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(visible = True),
        type='date'
    )
)

# creating figure with the both traces and layout
fig = dict(data= [trace], layout=layout)

#rendering the graphs
iplot(fig) #it's an equivalent to plt.show()
```
